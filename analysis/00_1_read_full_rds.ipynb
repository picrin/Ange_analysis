# Step 0 Part 1 -- Converting full count matrix

The data we get for the entire sample is in an `.rds` format. This is R-specific, we will convert it to a format that we can read in with Python. This notebook should be run on all samples (in this case `control` and `treated`) to get the appropriate full count matrices.

```r
sample <- "control"
```

```r
`+` <- function(e1, e2) {
  if (is.character(e1) | is.character(e2)) {
    paste0(e1, e2)
  } else {
    base::`+`(e1, e2)
  }
}
```

```r
dat <- readRDS("source/" + sample + ".rds")
```

```r
dat$cm
```

```r
write(x = rownames(dat$cm), file = "source/" + sample + ".all.genes.tsv")
write(x = colnames(dat$cm), file = "source/" + sample + ".all.cells.tsv")
```

```r
writeMM(dat$cm, file = "source/" + sample + ".all.mm")
```
