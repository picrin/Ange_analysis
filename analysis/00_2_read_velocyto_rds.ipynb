# Step 0 Part 2 -- Converting velocyto count matrix

As part of the velocyto data generation, we also get an `.rds` file with three count matrices, one each for categories `intron`, `exon`, and `spanning`. Again, this is R-specific, we will convert it to a format that we can read in with Python. This notebook should be run on all samples (in this case `control` and `treated`) to get the appropriate partial count matrices.

```r
sample <- "treated"
```

```r
`+` <- function(e1, e2) {
  if (is.character(e1) | is.character(e2)) {
    paste0(e1, e2)
  } else {
    base::`+`(e1, e2)
  }
}
```

```r
dat <- readRDS("source/" + sample + ".matrices.rds")
```

```r
dat$intron
```

```r
write(x = rownames(dat$intron), file = "source/" + sample + ".intron.genes.tsv")
write(x = colnames(dat$intron), file = "source/" + sample + ".intron.cells.tsv")
```

```r
write(x = rownames(dat$exon), file = "source/" + sample + ".exon.genes.tsv")
write(x = colnames(dat$exon), file = "source/" + sample + ".exon.cells.tsv")
```

```r
write(x = rownames(dat$spanning), file = "source/" + sample + ".spanning.genes.tsv")
write(x = colnames(dat$spanning), file = "source/" + sample + ".spanning.cells.tsv")
```

```r
writeMM(dat$intron, file = "source/" + sample + ".intron.mm")
```

```r
writeMM(dat$exon, file = "source/" + sample + ".exon.mm")
```
