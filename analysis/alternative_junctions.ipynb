```python
import os
import re
from gtfparse import read_gtf
import pandas as pd
pd.options.display.max_columns = None
```

```python
genecode_file = "gencode.v29.primary_assembly.annotation.gtf"
```

```python
df = read_gtf(genecode_file)
```

```python
df.query("gene_name == 'TIMP1' and feature == 'transcript'")
```

```python
gene = df.query("gene_name == 'SERPINE1'")
```

```python
gene
```

```python
transcripts = []
current_transcript = []
for row in gene.itertuples():
    if row.feature == "gene":
        continue
    if row.feature == "transcript":
        if current_transcript:
            transcripts.append(current_transcript)
        current_transcript = []
    if row.feature == "CDS":
        current_transcript.append((row.start, row.end))
if current_transcript:
    transcripts.append(current_transcript)
```

```python
junctions = []
```

```python
transcripts
```

```python
for transcript in transcripts:
    for i in range(len(transcript) - 1):
        junctions.append((transcript[i][1] + 1, transcript[i + 1][0] - 1))
```

```python
import collections
gene_bam = pysam.AlignmentFile("serpine1.sorted.bam", "rb")
bam_junctions = collections.Counter()
for read in gene_bam:
    #print({i: v for i, v in read.tags})
    
    junctions = {i: v for i, v in read.tags}['jI']
    if len(junctions) % 2 == 0:
        for junction in list(zip(junctions[::2], junctions[1::2])):
            bam_junctions[junction] += 1
    #    if junction != -1:
    #        print(junction)
    #        bam_junctions[junction] += 1
gene_bam.close()
bam_junctions
```
